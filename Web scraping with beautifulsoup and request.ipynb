```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
print(soup.prettify())
```

```python
soup.find_all('table')
```

```python
soup.find('table', class_ = "wikitable sortable sticky-header-multi sort-under")
```

```python
soup.find_all('th', scope = 'col')
```

```python
column_names_all = soup.find_all('th', scope='col')
column_names_all
```

```python
column_names = [names.text.strip() for names in column_names_all]
column_names
```

```python
column_names = column_names[:7]
column_names
```

```python
column_names[6]
```

```python
temp = column_names[6]
temp = temp[:11]
column_names[6] = temp
column_names[6]
```

```python
column_names
```

```python
import pandas as pd
df = pd.DataFrame(columns = column_names)
```

```python
df
```

```python
table = soup.find_all('table')[0]
table
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
    print(row.find_all('td'))
```

```python
for row in column_data[2:]:
    row_data = row.find_all('td')
    each_row_data = [data.text.strip() for data in row_data[:6]]
    print(each_row_data)
    length = len(df)
    df.loc[length] = each_row_data
```

```python
df
```

```python
df = pd.DataFrame(columns=column_names)
df
```

```python
df.drop('Rank', axis='columns', inplace=True)
df
```
